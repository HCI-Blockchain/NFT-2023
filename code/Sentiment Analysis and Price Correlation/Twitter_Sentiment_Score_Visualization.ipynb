# Data Visualization

```python
fig1 = make_subplots()
fig1.add_trace(go.Scatter(x=score17_18['day'], y=score17_18['sentiment_pos'],
              mode='lines', name='Positive'))
fig1.add_trace(go.Scatter(x=score17_18['day'], y=score17_18['sentiment_neg'],
              mode='lines', name='Negative'))
fig1.update_layout(title='2017-2018 Twitter Sentiment Score' ,autosize=True,)

py.iplot(fig1)

fig2 = make_subplots()
fig2.add_trace(go.Scatter(x=score18_19['day'], y=score18_19['sentiment_pos'],
              mode='lines', name='Positive'))
fig2.add_trace(go.Scatter(x=score18_19['day'], y=score18_19['sentiment_neg'],
              mode='lines', name='Negative'))
fig2.update_layout(title='2018-2019 Twitter Sentiment Score' ,autosize=True,)

py.iplot(fig2)

fig3 = make_subplots()
fig3.add_trace(go.Scatter(x=score19_20['day'], y=score19_20['sentiment_pos'],
              mode='lines', name='Positive'))
fig3.add_trace(go.Scatter(x=score19_20['day'], y=score19_20['sentiment_neg'],
              mode='lines', name='Negative'))
fig3.update_layout(title='2019-2020 Twitter Sentiment Score' ,autosize=True,)

py.iplot(fig3)

fig4 = make_subplots()
fig4.add_trace(go.Scatter(x=score20_21['day'], y=score20_21['sentiment_pos'],
              mode='lines', name='Positive'))
fig4.add_trace(go.Scatter(x=score20_21['day'], y=score20_21['sentiment_neg'],
              mode='lines', name='Negative'))
fig4.update_layout(title='2020-2021 Twitter Sentiment Score' ,autosize=True,)

py.iplot(fig4)
```
