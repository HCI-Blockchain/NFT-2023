```python
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cryptopunks_gender_skintone_info_new.csv')
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'], utc=True)

df['date'] = df['date'].apply(lambda x: x.replace(tzinfo=None))

df['dte'] = pd.to_datetime(df['date'])

# set time column as index
df = df.set_index('date')

# sort dataframe by index (i.e., by time) in ascending order
df = df.sort_index()
```

```python
df = df.reset_index()
```

```python
df
```

```python
# drop rows with duplicate names
df = df.drop_duplicates(subset=['img_url'])

# display the updated dataframe
df
```

```python
# replace "Non-human" values in the "skin_tone" column with their corresponding value in the "type" column
df.loc[df['skin_tone'] == 'Non-human', 'skin_tone'] = df['type']

# print the modified dataframe
df.head()
```

```python
df_gs = df[['gender','skin_tone']]
df_gs.head()
```

```python
pivot_df = df_gs.pivot_table(index='gender', columns='skin_tone', aggfunc=len)
```

```python
# add a row and column for total numbers
pivot_df.loc['Total'] = pivot_df.sum()
pivot_df['Total'] = pivot_df.sum(axis=1)

# add a row and column for proportion of each gender and skin tone
total_num = pivot_df.loc['Total', 'Total']
```

```python
pivot_df = pivot_df.fillna(0)
```

```python
# calculate the total number of observations
total = 6275

# calculate the proportion of each cell
pivot_df_prop = pivot_df / total * 100
```

```python
pivot_df 
```

```python
# create the heatmap graph
fig = px.imshow(pivot_df_prop.values,
                labels=dict(x="Skin Tone", y="Gender"),
                x=list(pivot_df_prop.columns),
                y=list(pivot_df_prop.index),
                color_continuous_scale='blues')

# add the corresponding values to each cell of the heatmap
for i in range(len(pivot_df_prop.index)):
    for j in range(len(pivot_df_prop.columns)):
        value = pivot_df_prop.values[i][j]
        count = pivot_df.values[i][j]
        font_size = 14
        font_color = 'black'
        if count > 2000:
            font_color = 'white'
        fig.add_annotation(x=j, y=i, text=f"{format(int(count), ',')}<br>({value:.1f}%)",
                           font=dict(size=font_size, color=font_color),
                           showarrow=False)

# add a title to the graph
fig.update_layout(title='CryptoPunks Gender and Skin Tone Distribution Heatmap')

# show the graph
fig.show()
```
